```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the colored image
image_path = '/content/duck.jpg'  # Replace with your image file path
colored_image = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(colored_image, cv2.COLOR_BGR2GRAY)

# Invert the grayscale image
inverted_gray_image = cv2.bitwise_not(gray_image)

# Apply Gaussian blur to the inverted grayscale image
blurred_image = cv2.GaussianBlur(inverted_gray_image, (111, 111), 0)

# Invert the blurred image to create a pencil sketch effect
inverted_blurred_image = cv2.bitwise_not(blurred_image)

# Create the pencil sketch by blending the grayscale image with the inverted blurred image
pencil_sketch = cv2.divide(gray_image, inverted_blurred_image, scale=256.0)

# Display the original colored image
plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(colored_image, cv2.COLOR_BGR2RGB))
plt.title("Original Colored Image")
plt.axis("off")

# Display the pencil sketch image
plt.subplot(1, 2, 2)
plt.imshow(pencil_sketch, cmap='gray')
plt.title("Pencil Sketch Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
